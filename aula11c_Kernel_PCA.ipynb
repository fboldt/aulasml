```python
from sklearn.datasets import make_swiss_roll

X_swiss, y_swiss = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
```

```python
from matplotlib.colors import ListedColormap
import numpy as np
darker_hot = ListedColormap(plt.cm.hot(np.linspace(0, 1, 256)))
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 5))
ax1 = fig.add_subplot(121, projection="3d")
ax1.scatter(X_swiss[:, 0], X_swiss[:, 1], X_swiss[:, 2], c=y_swiss, cmap=darker_hot)
ax1.view_init(10, -70)
plt.show()
```

```python
from sklearn.decomposition import KernelPCA

lin_pca = KernelPCA(kernel="linear")
rbf_pca = KernelPCA(kernel="rbf")
cos_pca = KernelPCA(kernel="cosine")

kernel_pcas = ((lin_pca, "Linear kernel"),
               (rbf_pca, "RBF kernel"),
               (cos_pca, "Cosine kernel"))

plt.figure(figsize=(11, 4))
for i, (kernel_pca, title) in enumerate(kernel_pcas):
    X_kernel_pca = kernel_pca.fit_transform(X_swiss)
    plt.subplot(1, 3, i + 1)
    plt.scatter(X_kernel_pca[:, 0], X_kernel_pca[:, 1], c=y_swiss, cmap=darker_hot)
    plt.title(title)
plt.show()
```
