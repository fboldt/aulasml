```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
print(data.keys())
```

```python
print(data.DESCR)
```

```python
X = data.data
print(X.shape)
```

```python
y = data.target
print(y.shape)
```

```python
import matplotlib.pyplot as plt
for i in [2]: #range(X.shape[1]):
    print(i)
    plt.scatter(X[:,i],y)
    plt.xlabel(data.feature_names[i])
    plt.ylabel("target")
    plt.show()
```

```python
print(X[:,2].shape)
```

```python
X2 = X[:,2:3]
print(X2.shape)
```

```python
def modelo_linear(X, a, b):
  y = a*X + b
  return y
```

```python
def plot_results(X, y, ypred):
  plt.scatter(X, y)
  plt.plot(X, ypred, 'r')
  plt.show()
```

```python
ypred = modelo_linear(X2, 1000, 100)
plot_results(X2, y, ypred)
```

```python
ypred = modelo_linear(X2, 1500, 150)
plot_results(X2, y, ypred)
```

```python
import numpy as np
```

```python
# mean absolute error
def mae(y, ypred):
  diff = y - ypred.reshape(len(y), )
  abs_diff = abs(diff)
  sum_error = np.sum(abs_diff)
  mae = sum_error / len(y)
  return mae

mae(y, modelo_linear(X2, 1000, 100)), mae(y, modelo_linear(X2, 1500, 150))
```

```python
# mean squared erorr
def mse(y, ypred):
  diff = y - ypred.reshape(len(y), )
  squared = diff**2
  mean = sum(squared)/len(squared)
  return mean

mse(y, modelo_linear(X2, 1000, 100)), mse(y, modelo_linear(X2, 1500, 150))
```

```python
# root mean squared error
def rmse(y, ypred):
  return np.sqrt(mse(y, ypred))

rmse(y, modelo_linear(X2, 1000, 100)), rmse(y, modelo_linear(X2, 1500, 150))
```
